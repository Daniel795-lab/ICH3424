## Modelo de Dally & Brown (1995)

Este modelo considera el área del roller y está basado en un balance de energía promediado en el tiempo e integrado en profundidad (es decir, unidimensional) que contiene contribuciones tanto del movimiento ondulatorio organizado (ola) como del roller, y se puede expresar como: 

$$\frac{d \overline{F_{W}}}{d x}+\frac{d \overline{F_{r}}}{d x}=-\overline{D}$$

$F_{W}$ es el flujo de energía asociado con el movimiento ondulatorio organizado, $F_{r}$ es el flujo de energía asociado con el roller y $D$ es la tasa de disipación de energía por unidad de área. La barra superior denota el promedio temporal durante un período de ola, y $x$ se dirige hacia la costa.

El flujo instantáneo de energía cinética a través de una sección vertical a medida que pasa el roller está dado por:

$$F_{r}= \mathrm{KE}_{r} C d_{r}$$

donde $C$ es la celeridad de la onda, $d_{r}$ es el espesor instantáneo del roller y $\mathrm{KE}_{r}$ es la densidad de energía cinética instantánea promediada sobre el espesor del roller. Al promediar en el tiempo, estas cantidades se pueden representar como:

$$\overline{\mathrm{KE}_{r}}=\frac{1}{2} \rho_{r}\left(\beta_{c} C\right)^{2}$$

$$\overline{d_{R}}=\frac{A}{CT}$$

$\rho_{r}$ es la densidad de masa promedio en el roller aireado (asumido constante), y $T$ es el periodo de onda. Se asume que la velocidad promedio en el roller es igual a la celeridad de la onda (Svendsen 1984a), el coeficiente $\beta_{c}$ es introducido para explicar la dependencia no lineal de la energía cinética de la velocidad instantánea de la partícula de agua. Sin embargo, este coeficiente es esperado a ser del orden de 1.0. El flujo de energía promediado en el tiempo asociado al roller se vuelve: 

$$\overline{F_{r}}=\frac{1}{2} \rho_{r}\left(\beta_{c} C\right)^{2} \frac{A}{T}$$

### Disipación

Aquí se adopta un término de disipación que se basa en el supuesto de que la energía se disipa en la interfaz entre el roller turbulento y el flujo organizado subyacente (Duncan, 1981).

El peso del roller por unidad de ancho de cresta ($W$) se puede expresar como:

$$W = \rho_{r} g A$$

Para que exista un equilibrio de fuerzas, el corte entre el roller y el fluido subyacente es:

$$S=W \sin \theta$$

Al proyectar este cizallamiento ligeramente inclinado, que se mueve con la celeridad de la onda, en la dirección del movimiento del roller, la tasa de disipación de energía promediada en el tiempo por unidad de área viene dada por:

$$\overline{D}=\frac{S \cos \theta \, C}{L}=\frac{\rho_{r} g A \sin \theta \cos \theta}{T}$$

Finalmente, la ecuación de gobierno para la evolución del roller es:

$$\frac{d \overline{F_{W}}}{dx}+\frac{d}{dx}\left(\frac{1}{2} \rho_{r} \beta_{c}^{2} C^{2} \frac{A}{T}\right)=-\rho_{r} g \beta_{D} \frac{A}{T}$$

donde $\beta_{D}$ es un coeficiente de disipación dado por el ángulo de inclinación del roller ($\sin \theta \cos \theta$) y es el coeficiente de calibración principal del modelo. En este modelo se intenta relacionar el área del roller con la altura de la ola local, lo que sería físicamente poco realista en la fase de crecimiento del roller.

### Solución numérica para le evolución del roller

Suponiendo que en el flujo de energía $\overline{F_{W}}$ y $C$ se conocen (o que la altura de onda y el nivel medio del agua se sabe que $\overline{F_{W}}$ y $C$ se pueden calcular a partir de la teoría lineal de ondas), la integración directa entre el punto $i$ y el punto $i + 1$ produce:

$$\overline{F_{W_{i+1}}}-\overline{F_{W_{i}}}+\frac{\rho_{r} \beta_{c}^{2}}{2 T}\left[\left(C^{2} A\right)_{i+1}-\left(C^{2} A\right)_{i}\right]=-\frac{\rho_{r} g \beta_{D}}{T} \underbrace{ \int_{x_{i}}^{x_{i+1}} A dx }_{\Delta x\left(\frac{A_{i+1}+A_{i}}{2}\right)}$$

### Corriente Undertow

Con una línea de coste presente, el flujo de volumen de agua en tierra que tiene lugar por encima del nivel medio del agua (MWL) debido al movimiento de las olas debe equilibrarse localmente con la descarga offshore por debajo de MWL debido a la corriente crossshore denominada Undertow (Dally, 1980). Esto se basa en la ecuación de continuidad promediada en el tiempo, integrada verticalmente:

$$\frac{d}{d x}[U(h+\bar{\eta})]+\frac{d Q_{W}}{d x}+\frac{d Q_{r}}{d x}=0$$

donde $U$ es la corriente promediada en la vertical (por debajo de MWL), $Q_{W}$ es el flujo de volumen por unidad de ancho de cresta asociado con el movimiento de onda organizado y $Q_{r}$ es el flujo de volumen de fluido debido al rodillo ($\rho_{r}A/\rho T$).

El balance de momentum crossshore promediado en el tiempo e integrado verticalmente, viene dado por:

$$\rho g(h+\bar{\eta}) \frac{d \bar{\eta}}{d x}+\frac{d S_{x x}}{d x}-\frac{d}{d x}\left[\rho U^{2}(h+\bar{\eta})\right]+\frac{d M_{r}}{d x}=-\bar{\tau}_{b}$$

Aquí, $S_{xx}$ es el tensor de radiación asociado al movimiento organizado de la ola (Longuet-Higgins & Stewart, 1964), $M_{r}$ es el flujo de momentum debido al roller (calculado como $\frac{\rho_{r}\beta_{c}CA}{T}$), y $\bar{\tau}_{b}$ es la tensión de corte promediada en el fondo promediada en el tiempo. El tercer término del lado izquierdo es la aceleración convectiva de la corriente.

La tensión del lecho se modela de acuerdo con el modelo de fricción cuadrática dado por:

$$\bar{\tau}_{b}=\rho(f / 8) \overline{u_{b}\left|u_{b}\right|}$$

Donde $f$ es el factor de fricción de Darcy-Weisbach y $u$ es la velocidad instantánea total (oscilatoria más media) cerca del lecho. Suponiendo un movimiento oscilatorio sinusoidal y asumiendo que la corriente media es pequeña, se puede demostrar que el esfuerzo de corte en el fondo se reduce a:

$$\bar{\tau}_{b}=(\rho f / 2 \pi) \hat{u}_{m} U$$

Donde $\hat{u}_{m}$ es la máxima velocidad oscilatoria (horizontal) cerca del fondo. Tomada desde la teoría lineal:

$$\hat{u}_{m} = \frac{gHT}{2L\cosh(2\pi D/L)}$$

El factor de fricción puede ser expresado como (Smith et al, 1993):

$$f = \frac{8gn^{2}}{(h+\bar{\eta})^{1/3}}$$

Con $n$ el coeficiente de Manning. 

Aplicando la condicion de borde que en la línea de costa no hay flujo medio:

$$U_{i+1}=-\left(Q_{W}+Q_{r}\right)_{i+1} /(h+\bar{\eta})_{i+1}$$

Una propiedad importante es:

$$(h+\bar{\eta}) \frac{d \bar{\eta}}{d x}=\frac{1}{2} \frac{d}{d x}(h+\bar{\eta})^{2}-(h+\bar{\eta}) \frac{d h}{d x}$$

Discretizando la ecuación de momentum:

$$\frac{\rho g}{2}\left[(h+\bar{\eta})_{i+1}^{2}-(h+\bar{\eta})_{i}^{2}\right]-\rho g \int_{x_{i}}^{x_{i+1}}\left[(h+\bar{\eta}) \frac{dh}{dx}\right] dx + \left(S_{xx_{i+1}}-S_{xx_{i}}\right)+\left(M_{r_{i+1}}-M_{r_{i}}\right) -\rho\left(\left[U^{2}(h+\bar{\eta})\right]_{i+1} - \left[U^{2}(h+\bar{\eta})\right]_{i}\right) = -\frac{\rho}{2\pi} \int_{x_{i}}^{x_{i+1}}\left(f \hat{u}_{m} U\right) dx $$

Utilizando la regla del trapecio para aproximar las integrales, tenemos:

$$ \frac{\rho g}{2}\left[(h+\bar{\eta})_{i+1}^{2}-(h+\bar{\eta})_{i}^{2}\right] -\rho g \Delta x\left(\frac{[(h+\bar{\eta})(-m)]_{i+1}+[(h+\bar{\eta})(-m)]_{i}}{2}\right) +\left(S_{xx_{i+1}}-S_{xx_{i}}\right)+\left(M_{r_{i+1}}-M_{r_{i}}\right)-\rho\left(\left[U^{2}(h+\bar{\eta})\right]_{i+1} + \left[U^{2}(h+\bar{\eta})\right]_{i}\right) = -\frac{\rho \Delta x}{2\pi}\left[\frac{\left(f \hat{u}_{m} U\right)_{i+1}+\left(f \hat{u}_{m} U\right)_{i}}{2}\right] $$

Aquí, $m$ corresponde a la pendiente local.

```
def crossU(Q_w,Q_r,D):
  U = -(Q_w+Q_r)/D
  return U

def bathymetry(XX,X,h_w,h_min):
  from scipy.interpolate import UnivariateSpline
  spline = UnivariateSpline(X, h_w)
  h_i = spline(XX)
  #from scipy import interpolate
  #tck = interpolate.splrep(X, h_w, s=0)
  #h_i = interpolate.splev(XX, tck, der=0)
  h = h_i - h_min
  return h

def dally_flux(H_2,H_1,D_1,D_2,C_1,C_2,n_1,n_2,B,gamma_d,gamma,K,dx,rho,g):
  E_w1 = B*rho*g*(H_1**2)
  Cg_1 = C_1*n_1
  F_w1 = E_w1*Cg_1
  E_w2 = B*rho*g*(H_2**2)
  Cg_2 = C_2*n_2
  F_w2 = E_w2*Cg_2
  H_s1 = gamma*D_1
  E_s1 = B*rho*g*(H_s1**2)
  F_s1 = E_s1*Cg_1
  H_s2 = gamma*D_2
  E_s2 = B*rho*g*(H_s2**2)
  F_s2 = E_s2*Cg_2

  if ((H_1/D_1)<gamma_d) or ( ((H_1/D_1)>=gamma_d) and ((F_w1-F_s1)<0)):
    a = 0
  else:
    a = 1

  return (F_w2 - F_w1) + (a*K*dx/2)*( ((F_w2-F_s2)/D_2) + ((F_w1-F_s1)/D_1) )

def roller_eq(A_2,A_1,C_1,C_2,F_w1,F_w2,T,beta_c,beta_d,rho_r,dx,g):
  return F_w2-F_w1 + (rho_r*(beta_c**2))/(2*T)*(C_2**2*A_2 - C_1**2*A_1) + (rho_r*g*beta_d*dx)/T*(A_2+A_1)/2

def momentum(D_2,D_1,U_1,U_2,S_xx1,S_xx2,u_m1,u_m2,m_1,m_2,f_1,f_2,rho,dx,g):
  tau_1 = (rho*dx*f_1*u_m1/(2*np.pi))*U_1
  tau_2 = (rho*dx*f_2*u_m2/(2*np.pi))*U_2
  return (rho*g*0.5)*((D_2**2)-(D_1**2))-(rho*g*dx*0.5)*(D_2*(-m_2)+D_1*(-m_1))+(S_xx2-S_xx1)-rho*((U_2**2)*D_2-(U_1**2)*D_1)+(tau_1+tau_2)/2

def momentum_roller(D_2,D_1,U_1,U_2,S_xx1,S_xx2,u_m1,u_m2,m_1,m_2,f_1,f_2,rho,dx,g,rho_r,beta_c,C_2,C_1,A_1,A_2,T):
  tau_1 = (rho*dx*f_1*u_m1/(2*np.pi))*U_1
  tau_2 = (rho*dx*f_2*u_m2/(2*np.pi))*U_2
  return (rho*g*0.5)*((D_2**2)-(D_1**2))-(rho*g*dx*0.5)*(D_2*(-m_2)+D_1*(-m_1))+(S_xx2-S_xx1)-rho*((U_2**2)*D_2-(U_1**2)*D_1)+ (rho_r*beta_c/T)*(C_2*A_2-C_1*A_1) +(tau_1+tau_2)/2
```

```
def cross_shore(X,Zf,H_wl,T,B,nseg,roller):
  # Librerias utilizadas
  import numpy as np
  from scipy.optimize import fsolve
  from scipy import interpolate
  
  # Parametros del modelo
  g       = 9.81 # Aceleración de gravedad [m/s^2]
  rho     = 1000 # Masa específica del agua [kg/m^3]
  rho_r   = 1000 # Masa específica del roller [kg/m^3]
  beta_c  = 1.0  # Parámetro de de velocidad en el roller (u_r = beta_c*c)
  beta_d  = 0.10 # Parametro para friccion entre escurrimiento organizado y roller (angulo de reposo)
  gamma_d = 0.8  # Índice para el inicio del rompimiento (H_b/h_b)
  w       = 2*np.pi/T # Frecuencia angular del oleaje [rad/s]
  manning = 0.02 # Coeficiente de fricción
  
  # Modelo de evolución de Dally, Dean & Dalrymple (1985)
  gamma = 0.5
  K     = 0.15
  
  # Discretización numérica
  tol      = 1e-4 # Tolerancia para el erroe relativo en iteraciones
  max_iter = 10   # Número de interaciones máximas
  h_w      = -Zf
  h_min    = 0 #Profundidad de agua mínima para efectuar los cálculos
  # Límites del dominio cross-shore
  x_min = fsolve(bathymetry,np.min(X),args=(X,h_w,h_min))
  x_max = np.max(X)
  # Dominio discreto
  dx  = (x_max-x_min)/nseg           # Paso de discretización
  X_d  = np.arange(x_min,x_max+dx,dx) # vector de coordenadas nodales
  m   = len(X_d)
  tck = interpolate.splrep(X, Zf, s=0)
  Zfd = interpolate.splev(X_d, tck, der=0)# vector de cotas de fondo en nodos
  h_d = -Zfd # vector de profundidades de agua en reposo
  m_d = np.gradient(h_d,X_d) # Pendiente local del fondo

  # Iniciación de variables para el cálculo
  A_0 = 0 # Area del roller off-shore
  h_0 = h_d[m-1]
  D_0 = h_0 # Profundidad media de agua off-shore
  H_0 = H_wl # Altura de onda off-shore
  C_0 = (g*D_0)**0.5 # Celeridad de la onda off-shore (aguas someras)
  Q_wx0 = B*C_0*(H_0**2)/D_0 # Flujo volumétrico debido al oleaje off-shore
  Q_rx0 = (rho_r*A_0)/(rho*T) # Flujo volumétrico debido al roller off-shore
  U_0   = crossU(Q_wx0,Q_rx0,D_0) # Corriente undertow off-shore
  gamma_d0 = gamma_d # Índice de rompimiento
  
  H_d = np.zeros(m)
  A_d = np.zeros(m)
  D_d = np.zeros(m)
  U_d = np.zeros(m)
  C_d = np.zeros(m)
  S_xxd = np.zeros(m)
  Fr_d  = np.zeros(m)
  U_md  = np.zeros(m)
  
  H_d[m-1] = H_0
  A_d[m-1] = A_0
  D_d[m-1] = D_0
  U_d[m-1] = U_0
  C_d[m-1] = C_0
  
  # Cálculo de propiedades hidrodinámicas
  for i in range(m-1,0,-1): #range(m-1,1,-1)
    # Propiedades integrales en el nodo i
    H_1  = H_d[i]
    A_1  = A_d[i]
    h_1  = h_d[i]
    D_1  = D_d[i]
    U_1  = U_d[i]
    m_1  = m_d[i]
    E_w1 = B*rho*g*(H_1**2) # Energia de la onda
    C_1  = (g*D_1)**0.5 # Celeridad de la onda (aguas someras)
    L_1  = C_1*T # longitud de la onda (aguas someras)
    k_1  = 2*np.pi/L_1 # Número de onda
    n_1  = 0.5*(1+(2*k_1*D_1)/(np.sinh(2*k_1*D_1))) # Coeficiente para flujo de energia (velocidad de grupo)
    E_r1 = (1/2)*rho_r*beta_c*C_1*A_1/T # Energia del roller
    F_wx1 = E_w1*C_1*n_1 #Flujo de energía debida al oleaje
    S_xx1 = (2*n_1-(1/2))*E_w1+2*E_r1 # Tensor de radiación incluyendo oleaje y roller
    u_m1  = (g*H_1*T)/(2*L_1*np.cosh(2*np.pi*D_1/L_1)) # Velocidad orbital en el fondo
    f_1   = (8*g*(manning**2))/(D_1**(1/3)) # Factor de friccion
    
    S_xxd[i] = S_xx1
    Fr_d[i]  = f_1
    U_md[i]  = u_m1
    
    # Iteraciones para resolucion de ecuacion no lineal
    err  = 1
    iter = 0
    H_2  = H_1
    A_2  = A_1
    D_2  = D_1
    U_2  = U_1
    
    # Evolucion de altura de ola (Modelo de Dally et al. JGR Vol. 90, No. C6, pp. 11917-11927, 1985)
    if (H_1/D_1) > gamma_d0:
      gamma_d0 = 0
    
    while (err > tol) and (iter < max_iter):
      # Propiedades integrales en el nodo i -1
      h_2   = h_d[i-1]
      m_2   = m_d[i-1]
      C_2   = (g*D_2)**0.5 # Celeridad (aguas someras)
      L_2   = C_2*T # longitud de la onda (aguas someras)
      k_2   = 2*np.pi/L_2
      u_m2  = (g*H_2*T)/(2*L_2*np.cosh(2*np.pi*D_2/L_2)) # Velocidad orbital en el fondo
      f_2   = (8*g*(manning**2))/(D_2**(1/3)) # Factor de fricción
      Q_wx2 = B*C_2*(H_2**2)/D_2 # Flujo volumétrico debido al oleaje
      n_2   = 0.5*(1+(2*k_2*D_2)/(np.sinh(2*k_2*D_2))) # Coeficiente para flujo de energia (velocidad de grupo)
      
      # Ecuación de conservación de energía para el oleaje
      # Modelo de Dally et al. (1985)
      H_2n = fsolve(dally_flux,H_2,args=(H_1,D_1,D_2,C_1,C_2,n_1,n_2,B,gamma_d0,gamma,K,dx,rho,g))
      
      if H_2n == 0:
        err_1 = tol/10
      else:
        err_1 = np.abs((H_2n-H_2)/H_2n)
      
      H_2   = H_2n # Altura de la onda (Corregida)
      E_w2  = B*rho*g*(H_2**2) # Flujo de energía (Corregido)
      F_wx2 = E_w2*C_2*n_2 # Flujo de energía (Corregido)
      u_m2  = (g*H_2*T)/(2*L_2*np.cosh(2*np.pi*D_2/L_2)) # Velocidad orbital en el fondo (Corregido)
      
      # Evolucion del Area del roller
      if roller == 1:
        A_2n = fsolve(roller_eq,A_2,args=(A_1,C_1,C_2,F_wx1,F_wx2,T,beta_c,beta_d,rho_r,dx,g))
      else:
        A_2n = 0
      
      if A_2n < 0:
        A_2n = 0
      
      if A_2n == 0:
        err_2 = tol/10
      else:
        err_2 = np.abs((A_2n-A_2)/A_2n)
      
      A_2   = A_2n 
      E_r2  = (1/2)*rho_r*beta_c*C_2*A_2/T # Energia del roller
      S_xx2 = (2*n_2-(1/2))*E_w2+2*E_r2 # Tensor de radiacion incluyendo el roller
      Q_rx2 = (rho_r*A_2)/(rho*T) # Flujo volumétrico debido al roller
      
      # Resolución de la ecuación de cantidad de movimiento
      D_2n  = fsolve(momentum,D_2,args=(D_1,U_1,U_2,S_xx1,S_xx2,u_m1,u_m2,m_1,m_2,f_1,f_2,rho,dx,g))
      #D_2n  = fsolve(momentum_roller,D_2,args=(D_1,U_1,U_2,S_xx1,S_xx2,u_m1,u_m2,m_1,m_2,f_1,f_2,rho,dx,g,rho_r,beta_c,C_2,C_1,A_1,A_2,T))
      err_3 = np.abs((D_2n-D_2)/D_2n)
      D_2   = D_2n
      U_2   = crossU(Q_wx2,Q_rx2,D_2) # Corriente undertow
      err   = np.max([err_1, err_2, err_3])
      iter  = iter + 1
    
    H_d[i-1]   = H_2
    A_d[i-1]   = A_2
    D_d[i-1]   = D_2
    U_d[i-1]   = U_2
    C_d[i-1]   = C_2
    S_xxd[i-1] = S_xx2
    Fr_d[i-1]  = f_2
    U_md[i-1]  = u_m2
    
    if D_2 <= 0: #0.0001:
      ind = i
      break
    else:
      ind = i-1
  
  X_d   = X_d[ind:m]
  Zfd   = Zfd[ind:m]
  H_d   = H_d[ind:m]
  h_d   = h_d[ind:m]
  A_d   = A_d[ind:m]
  D_d   = D_d[ind:m]
  U_d   = U_d[ind:m]
  C_d   = C_d[ind:m]
  S_xxd = S_xxd[ind:m]
  Fr_d  = Fr_d[ind:m]
  U_md  = U_md[ind:m]
  
  return H_d,D_d,U_d,A_d,S_xxd,C_d,U_md,Fr_d,h_d,X_d,Zfd
```

```
import numpy as np
# Test Hansen and Svendsen (1984)

H_wl = 0.1540 # Altura de ondas off-shore
T   = 2 # Periodo del oleaje incidente
B   = 1/8 # Coeficiente de forma de onda
roller = 1 # roller=1 (incluye roller) roller=0 (sin roller)
nseg   = 50 # Numero de elementos discretizacion

# Batimetría
dx = 0.06
X  = np.arange(0,7.62+dx,dx)
Zf = -1/34*X # Playa de pendiente uniforme 1/34

# Mediciones experimentales

# H - alturas de olas medidas
Hm   = np.array([0.0510, 0.0670, 0.0800, 0.1100, 0.1650, 0.1730, 0.1540])
Xm_H = np.array([2.6000, 3.2200, 4.0800, 5.0800, 6.0900, 6.9200, 7.6200])
# Set-up medido
MWL    = np.array([0.0100, 0.0070, 0.0020, -0.0020, -0.0030])
Xm_MWL = np.array([2.6000, 3.2200, 4.0800,  5.0800,  6.0900])
# Undertow medido
Um   = np.array([-0.0630, -0.0730, -0.0870, -0.1020, -0.0440])
Xm_U = np.array([ 2.6000,  3.2200,  4.0800,  5.0800,  7.6200])

H_d,D_d,U_d,A_d,S_xxd,C_d,U_md,Fr_d,h_d,X_d,Zfd = cross_shore(X,Zf,H_wl,T,B,nseg,roller)

Hydrodynamics = np.array([H_d,A_d,D_d-h_d,U_d,Zfd])
```

```
import matplotlib.pyplot as plt

alto  = 7.5
ancho = 8.5
fig = plt.figure(figsize=(ancho, alto))

title    = 'Surf Zone Hydrodynamics'
labels   = ['$H$ (m)','$A$ (m$^{2}$)','$\\overline{\\eta}$ - MWL (m)','$U$ (m/s)','$h(x)}$ (m)']
colors   = ['b','gray','r','c','k']
lines    = ['solid','dashed','dashed','solid','solid']

# from Model
for i in range(5):
  plt.plot(X_d,Hydrodynamics[i,:],colors[i],linestyle=lines[i],label=labels[i])

# from observed data
plt.plot(Xm_H,Hm,'o',c=colors[0])
plt.plot(Xm_MWL,MWL,'x',c=colors[2])
plt.plot(Xm_U,Um,'s',c=colors[3])

plt.title(title)
plt.xlabel('Cross-shore distance (m)')
plt.ylim(np.min(Zf),1.1*np.max([H_d,A_d]))
plt.xlim(X_d[0],np.max(X_d))
plt.legend()
plt.grid(color='lightgray', linestyle='--', linewidth=0.25)
plt.show()
```
